# Summary

In this notebook, we demonstrate the identification of the tidal tails of the globular cluster Palomar 5 in the NSC DR1 catalog, as well as a join against Gaia DR2 to explore the proper motion of the cluster and its tails. We issue a query around the position of the globular cluster and retrieve all data from NSC DR1 within an area of sky, as well as information from Gaia DR2 for any objects that have matches. 

** Background **

Odenkirchen M. et al., 2001, ApJ, 548, L165 found extended tidal tails around the globular cluster Palomar 5 using data from SDSS.  The tails indicate that the cluster is in the process of evaporation.  They are useful tracers of the Milky Way's gravitational potential, and are potential probes of discrete dark matter subhalos in the Milky Way.

** Data retrieval **

We will use catalog data found in three tables: the NSC DR1 object table, the Gaia DR2-NSC DR1 crossmatch table, and the Gaia DR2 source table.

** Detection **

In order to pick out the tidal tails, we'll filter the NSC DR1 catalog results to isolate the Pal 5 sequence, and then run a differential convolution filter of the spatial distribution to bring out the tails.  We'll then explore the proper motions of the cluster and tails using the Gaia DR2 data.

### Query logic

The goal of the query is to retrieve all stars from NSC DR1 within a significant radius of Pal 5, in this case 2 degrees.  We also want to retrieve measurements from Gaia DR2 when it's available.  This kind of query is referred to as a LEFT JOIN: we want to get *all* of the data from one table (nsc_dr1) as well as all of the data that intersects (through a cross-match)with another table (gaia_dr2).

Rather than do the positional cross-match on the fly, we'll make use of a precomputed crossmatch table (gaia_dr2.nsc_dr1) which we need to INNER JOIN against the Gaia source table (gaia_dr2.gaia_source) in order to access the Gaia measurements.

The logic of the query then goes like this:
- JOIN the crossmatch table against the Gaia source table, subject to the spatial constraint of being within 2 degrees of Pal 5, and materialize this table in the query (we'll call it gt)
- Perform the LEFT JOIN of nsc_dr1 with this materialized table, and constrain nsc_dr1 to also only contain sources within 2 degrees of Pal 5
- Return rows from both nsc_dr1 and gaia_dr2.  If there are objects in nsc_dr1 that have no Gaia DR2 crossmatch, the Gaia columns for those objects will be null.


```python
%%time
radius=2.0
query2="""
SELECT n.id,n.ra,n.dec,n.pmra,n.pmdec,n.pmraerr,n.pmdecerr,n.mjd,n.umag,n.uerr,
        n.gmag,n.gerr,n.rmag,n.rerr,n.imag,n.ierr,n.zmag,n.zerr,gt.*
FROM nsc_dr1.object AS n
LEFT JOIN (SELECT x.source_id as gaia_id, x.id as nsc_id, g.phot_g_mean_mag as gaia_G,
      x.ra as gaia_ra, x.dec as gaia_dec, x.dist as dist,
      g.pmra as gaia_pmra, g.pmra_error as gaia_pmra_err, g.pmdec as gaia_pmdec, 
      g.pmdec_error as gaia_pmdec_err, g.parallax,g.parallax_error,g.bp_rp,
      g.astrometric_excess_noise,g.astrometric_excess_noise_sig, g.duplicated_source
    FROM gaia_dr2.gaia_source as g
    INNER JOIN gaia_dr2.nsc_dr1 as x
    ON g.source_id=x.source_id
    WHERE (q3c_radial_query(g.ra,g.dec,{0},{1},{2}))) as gt
ON n.id=gt.nsc_id
WHERE (q3c_radial_query(n.ra,n.dec,{0},{1},{2}))
""".format(c.ra.value,c.dec.value,radius)

try:
    result=qc.query(sql=query2)
except Exception as e:
    print(e.message)
```

# Spatial filtering
We'll see the tails more clearly if we apply a spatial filter, here the same Mexican hat filter that we used in the Finding Dwarfs notebook.  In the figure below, you can clearly see the S-shape of the cluster's tidal tails.

```python
## a convolution filter
def dwarf_filter (ra,dec,fwhm_small=2.0,fwhm_big=20):

    """Differential convolution with 2D Gaussian kernels.
    
       Based on Koposov et al. (2008).
       Code by Ken Mighell and Mike Fitzpatrick.
       Minor edits by RN.
       
       Parameters
       ----------
       ra, dec : float or array
           RA & Dec in degrees.
    
       fwhm_small, fwhm_big : float
           Full-width half maximum sizes of the small and big Gaussian kernels
           to use in convolution, in arcminutes.
    """
    
    x, y = ra, dec

    print("Computing differential convolution")

    # Information about declination (y) [degrees]
    ymean = (y.min() + y.max()) / 2.0
    ydiff_arcmin = (y.max() - y.min()) * 60.0 # convert from degrees to arcmin

    # Information about right ascension (x) [degrees in time]:
    xdiff = x.max() - x.min() # angular separation [degrees (time)] 
    xmean = (x.min() + x.max()) / 2.0

    # convert from degrees in time to separation in angular degrees:
    xdiff_angular = (x.max() - x.min()) * np.cos(ymean*(np.pi/180.0))

    # convert from degress to arcmin
    xdiff_angular_arcmin = xdiff_angular * 60.0 

    # Get the number of one-arcmin pixels in the X and Y directions:
    nx = np.rint(xdiff_angular_arcmin).astype('int')
    ny = np.rint(ydiff_arcmin).astype('int')

    # Create a two-dimensional histogram of the raw counts:
    Counts, xedges, yedges  = np.histogram2d (x, y, (nx,ny) )
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
    raw_hist = np.rot90(Counts).copy() # hack around Pythonic weirdness

    # Make the small and big Gaussian kernels with a standard deviation
    # of the given FWHM in arcmin^2 pixels.
    kernel_small = convolution.Gaussian2DKernel(fwhm_small/2.35,factor=1)
    kernel_big = convolution.Gaussian2DKernel(fwhm_big/2.35,factor=1)

    # Compute the differential convolution kernels.
    conv_big = convolution.convolve(raw_hist, kernel_big)
    conv_small = convolution.convolve(raw_hist, kernel_small)
    conv_delta = conv_small - conv_big
    delta = conv_delta.copy()

    # Compute statistics and the floor
    mean = np.mean(delta, dtype='float64')
    sigma = np.std(delta, dtype='float64')
    sigmaRaw = np.std(raw_hist,dtype='float64')
    median = np.median(delta) # not used
    floor = mean

    clipped = delta.copy()
    clipped[delta < floor] = floor

    # Return the computed fields.
    return raw_hist, extent, delta, clipped, sigma
```

```python
raw_hist, extent, delta, clipped, sigma = \
    dwarf_filter(dfp['ra'],dfp['dec'],fwhm_small=8.0,fwhm_big=120)
```

# Exploring the proper motions
Gaia DR2 has proper motion measurements for some of the objects in our dataframe.  We'll use matplotlib's quiver function to plot them.  In the plot below, you can tell that there is a bulk motion towards the southwest, likely the reflex motion of the Sun.  Pal 5's proper motion is smaller than that of the foreground screen.

```python
east_left=True
if east_left:
    flip=-1
else:
    flip=1
x0=dfp['ra']
y0=dfp['dec']
dx=dfp['gaia_pmra']
dy=dfp['gaia_pmdec']

fig=plt.figure(figsize=(10,10))
plt.quiver(x0,y0,flip*dx,dy,scale_units='width',scale=200)
plt.xlabel('RA')
plt.ylabel('Dec')
if east_left: plt.gca().invert_xaxis()
```

If we limit the plot to those objects with smaller propoer motions, Pal 5 and its tails are easier to see.

```python
pm_limit=6
pmp=np.sqrt(dfp['gaia_pmra']**2+dfp['gaia_pmdec']**2)
pmperr=np.sqrt(dfp['gaia_pmra_err']**2+dfp['gaia_pmdec_err']**2)
lpm=pmp<pm_limit
x0=dfp[lpm]['ra']
y0=dfp[lpm]['dec']
dx=dfp[lpm]['gaia_pmra']
dy=dfp[lpm]['gaia_pmdec']

fig=plt.figure(figsize=(10,10))
plt.quiver(x0,y0,flip*dx,dy,scale_units='width',scale=200)
plt.xlabel('RA')
plt.ylabel('Dec')
if east_left: plt.gca().invert_xaxis()
```

Let's take the mean Pal 5 proper motion and subtract it from our vectors, to view the proper motions in Pal 5's reference frame.  From the figure below, it looks like we might be able to cut on the proper motion s a bit further to isolate Pal 5 and its tails.

```python
core=(x0>228.9)&(x0<229.1)&(y0>-0.2)&(y0<0) # Pal 5 core
mpmra=np.mean(dx[core])
mpmdec=np.mean(dy[core])
dxp5=dx-mpmra
dyp5=dy-mpmdec
print(mpmra,mpmdec)
```

```python
fig=plt.figure(figsize=(10,10))
plt.quiver(x0,y0,flip*dxp5,dyp5,scale_units='width',scale=200)
plt.xlabel('RA')
plt.ylabel('Dec')
if east_left: plt.gca().invert_xaxis()
```

In the figure below, we'll only show points that have proper motions close to those of Pal 5 within some limit.  It removes a bit more of the contamination.

```python
lim=3. # try experimenting with this proper motion cut
p5=(np.abs(dxp5)<lim)&(np.abs(dyp5)<lim)
fig=plt.figure(figsize=(10,10))
plt.quiver(x0[p5],y0[p5],flip*dxp5[p5],dyp5[p5],scale_units='width',scale=200)
plt.xlabel('RA')
plt.ylabel('Dec')
if east_left: plt.gca().invert_xaxis()
```

Let's see what happens when we apply our spatial filter to this proper motion-selected sample.  The northern tail clearly stands out.
